# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['25'] = 'Midnight Picture Show'
```

```python
production_companies_overview_sum = {}
production_companies_overview_count = {}

for index, row in data.iterrows():
    for production_company in row['production_companies']:
        if production_company not in production_companies_overview_sum:
            production_companies_overview_sum[production_company] = len(row['overview'])
            production_companies_overview_count[production_company] = 1
        else:
            production_companies_overview_sum[production_company] += len(row['overview'])
            production_companies_overview_count[production_company] += 1
                
#         if len(row['overview']) >= 83:
#                print(row['overview'])
            
# dictionairy_sort_by_key(production_companies_overview_sum)[-9:]

production_companies_overview_sum

companies_my_sum = pd.DataFrame.from_dict(production_companies_overview_sum, 
                           orient='index',
                           columns=['sum'])

# print(companies_my_sum.count)

companies_my_count = pd.DataFrame.from_dict(production_companies_overview_count, 
                           orient='index',
                           columns=['count'])

# print(companies_my_count.count)

result = companies_my_sum.join(companies_my_count)
result

result['mean_length'] = result['sum'] / result['count']


result.query('mean_length >= ' + str(result['mean_length'].max()))
```
